```python
import soundfile as sf
import matplotlib.pyplot as plt
import numpy as np

wave, srate = sf.read("little_wing.wav")
wave = wave[45000:srate*2, 0]

sf.write('slice_little_wing.wav', wave, srate)

plt.plot(wave)
plt.show()

print(len(wave))

yf = np.fft.fft(wave)

# Transform again
yf_t = np.fft.fft(yf)

# Reverse and normalize
wave_new = np.divide(yf_t[-np.arange(yf_t.shape[0])],wave.shape[-1])


plt.plot(abs(yf[:7000]))
plt.show()

plt.plot(wave_new)
plt.show()
```
